```python
import kagglehub
from glob import glob
import os
import pandas as pd

path = kagglehub.dataset_download("aeryss/lte-dataset")
base_dir = os.path.join(path, 'Dataset')
file_list = []
for dir in os.listdir(base_dir):
    file_list += glob(os.path.join(os.path.join(base_dir, dir), "*.csv"))

for idx, filedir in enumerate(file_list):
    if idx ==0:
        df = pd.read_csv(filedir)
    else:
        temp = pd.read_csv(filedir)
        df = pd.concat([df, temp], axis = 0)

df["Fault"] = 0
df.loc[df['DL_bitrate'] < 3 * 1024, "Fault"] = 1
df.drop(columns = ['DL_bitrate', 'UL_bitrate'], inplace = True)
fin_df = df.loc[df['NetworkMode'] == "LTE", :].copy()
```

```python
with open('./test_idx.txt', 'r') as f:
  test_idx = f.read()
fin_df.reset_index(drop = True, inplace = True)
test_idx_list = [int(idx) for idx in test_idx.split(",")]
test_df = fin_df.copy().iloc[test_idx_list, :]
train_df = fin_df.copy().drop(index = test_idx_list)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
X = fin_df.drop('Fault', axis=1)
y = fin_df['Fault']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test)


print(accuracy_score(y_test, lr_pred))
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
gb_pred = gb.predict(X_test)


print(accuracy_score(y_test, gb_pred))
```
