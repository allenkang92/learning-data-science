```python
# pyplot으로 간단한 그래프를 그린다
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5) # -5부터 5까지
y = 2 * x # x에 2를 곱해서 y좌표로 한다

plt.plot(x, y)
plt.show()
```

```python
# 그래프를 꾸민다
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5) # 생성할 샘플 수는 지정하지 않으면 기본 값이 50
y_1 = 2 * x
y_2 = 3 * x

# 축의 레이블
plt.xlabel("x value", size = 14) # 축 레이블의 문자 크기를 14로 지정
plt.ylabel("y value", size = 14)

# 그래프의 타이틀
plt.title("My Graph")

# 그리드 표시
plt.grid()

# 플롯 시에 범례와 선의 스타일을 지정
plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2", linestyle="dashed")
plt.legend() # 범례 표시

plt.show()
```

```python
# scatter() 함수로 산포도를 표시한다
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1.2, 2.4, 0.0, 1.4, 1.5, 0.3, 0.7]) # x 좌표
y = np.array([2.4, 1.4, 1.0, 0.1, 1.7, 2.0, 0.6]) # y 좌표

plt.scatter(x, y) # 산포도의 플롯
plt.grid() # 격자선 표시
plt.show()
```

```python
# 히스토그램을 표시한다
import numpy as np
import matplotlib.pyplot as plt

data = np.array([0, 1, 1, 2, 2, 2, 3, 3, 4, 5, 6, 6, 7, 7, 7, 8, 8, 9])

plt.hist(data, bins=10) # 히스토그램 bins는 기둥의 수
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3) # x의 범위를 지정
y_1 = 1.5 * x # x에 연산을 시행하는 y_1으로 한다.
y_2 = -2 * x + 1 # x에 연산을 시행하는 y_2로 한다.

# 축의 레이블
plt.xlabel("x value", size = 14)
plt.ylabel("y value", size = 14)

# 그래프의 타이틀
plt.title("My Graph")

# 그리드 표시
plt.grid()

# 플롯 범례와 선의 스타일을 지정
plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2", linestyle="dashed")
plt.legend()    # 범례를 표시

plt.show()
```

사용 가능한 linestyle 옵션:
'solid' 또는 '-': 실선 (기본값)
'dashed' 또는 '--': 점선
'dashdot' 또는 '-.': 일점쇄선
'dotted' 또는 ':': 점선
