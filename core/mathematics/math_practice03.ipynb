거듭제곱과 제곱근은 수식을 기술하는데 중요함.


거듭제곱과 제곱근을 코드로 구현


```python
# 거듭제곱의 그래프를 그린다
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    a = 3
    return x ** a #x의 a 제곱

x = np.linspace(0, 2)
y = my_func(x) # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
# 제곱근의 그래프를 그린다
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x) # x의 양의 제곱근, x ** 1/2도 가능

x = np.linspace(0, 9)
y = my_func(x) # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return x ** 0.5 + 1

x = np.linspace(0, 4)
y = my_func(x) # y = f(x)

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
