삼각함수

```python
# sin()함수와 cos()함수를 그래프로 그린다
import numpy as np
import matplotlib.pyplot as plt

def my_sin(x):
    return np.sin(x) # sin(x)

def my_cos(x):
    return np.cos(x) # cos(x)

x = np.linspace(-np.pi, np.pi) # -pi부터 pi(라디안)까지
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid

plt.show()
```

```python
# tan()함수를 그래프로 그린다
import numpy as np
import matplotlib.pyplot as plt

def my_tan(x):
    return np.tan(x) # tan(x)

x = np.linspace(-1.3, 1.3) # -1.3부터 1.3(라디안)까지
y_tan = my_tan(x)

plt.plot(x, y_tan, label="tan")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def my_sin(x):
    return np.sin(x) # sin(x)

def my_cos(x):
    return np.cos(x) # cos(x)

x = np.linspace(-2*np.pi, 2*np.pi) # x의 범위를 지정
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()

plt.show()
```
