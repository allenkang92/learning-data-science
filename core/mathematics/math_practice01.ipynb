변수와 상수는 수식을 다루기 위해 기초가 되는 개념

간단히 말하면 변수 : 변화하는 수,
상수 : 일정한, 변화하지 않는 수

```python
# 변수와 상수를 사용해 직선을 그려보기
import numpy as np
import matplotlib.pyplot as plt

a = 1.5 # a : 상수
x = np.linspace(-1, 1) # x : 변수 -1부터 1의 범위
y = a * x # y : 변수

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

b = 3 # b : 상수
x = np.linspace(-1, 1) # x : 변수
y = b * x # y : 변수

plt.plot(x, y)
plt.xlabel("x", size = 14)
plt.ylabel("y", size = 14)
plt.grid()
plt.show()
```
