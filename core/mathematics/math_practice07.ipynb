난수, 규칙성이 없는 예측할 수 없는 수치(미확정 수치)

-> 인공지능에서는 파라미터의 초기화 등에 난수를 활용한다.

```python
# 1부터 6까지의 정수의 난수를 생성한다
import numpy as np

r_int = np.random.randint(6) + 1 # 0부터 5까지의 난수에 1을 더한다
print(r_int)
```

```python
# 0부터 1까지의 소수의 난수를 생성한다
import numpy as np

r_dec = np.random.rand() # 0부터 1 사이의 소수를 랜덤으로 반환한다
print(r_dec)
```

```python
# 난수의 균일한 분포
import numpy as np
import matplotlib.pyplot as plt

n = 1000 # 샘플 수
x = np.random.rand(n) # 0-1의 균일한 난수
y = np.random.rand(n) # 0-1의 균일한 난수

plt.scatter(x, y) # 산포도를 플롯
plt.grid()
plt.show()
```

```python
# 정규분포를 따르는 난수의 분포
import numpy as np
import matplotlib.pyplot as plt

n = 1000 # 샘플 수
x = np.random.randn(n) # 정규분포를 따른 난수의 분포
y = np.random.randn(n) # 정규분포를 따른 난수의 분포

plt.scatter(x, y) # 산포도의 플롯
plt.grid()
plt.show()
```
